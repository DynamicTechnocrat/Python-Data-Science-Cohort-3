```python
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Set a Seaborn style for better aesthetics
sns.set_style("whitegrid")

# Group and plot
sales_by_year = df.groupby("year")['Weekly_Sales'].sum()

# Create a figure and axis
plt.figure(figsize=(10, 6))
bars = plt.bar(sales_by_year.index, sales_by_year.values, color='skyblue', edgecolor='black')

# Title and labels
plt.title("Total Weekly Sales by Year", fontsize=16, fontweight='bold')
plt.xlabel("Year", fontsize=14)
plt.ylabel("Total Sales", fontsize=14)

# Add value labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1e6, f'{yval:,.0f}', ha='center', va='bottom', fontsize=10)

# Tidy up layout
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Prepare data
monthly_sales = df.groupby("month")["Weekly_Sales"].sum().sort_index()

# Aesthetic settings
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

# Bar plot
bars = plt.bar(monthly_sales.index, monthly_sales.values, color='skyblue', edgecolor='black', label='Total Sales')

# Line plot on top of bars
plt.plot(monthly_sales.index, monthly_sales.values, color='darkblue', marker='o', linewidth=2, label='Trend')

# Annotate bars with values
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1e6, f'{height:,.0f}', 
             ha='center', va='bottom', fontsize=9)

# Labels and title
plt.title("Total Weekly Sales by Month", fontsize=16, fontweight='bold')
plt.xlabel("Month", fontsize=14)
plt.ylabel("Total Sales", fontsize=14)
plt.xticks(monthly_sales.index, 
           ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
            'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], 
           fontsize=11)

# Legend
plt.legend()

# Layout and show
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style for better visuals
sns.set(style="white", font_scale=1.2)

# Create figure and axes
plt.figure(figsize=(15, 8))

# Create heatmap
heatmap = sns.heatmap(
    df.corr(),
    annot=True,
    fmt=".2f",                 # Format numbers to 2 decimal places
    cmap="coolwarm",           # Blue to red color map
    linewidths=0.5,            # Lines between cells
    linecolor='white',         # Color of those lines
    square=True,               # Make cells square
    cbar_kws={"shrink": 0.8}   # Shrink color bar slightly
)

# Title
plt.title("Correlation Heatmap", fontsize=16, fontweight='bold', pad=20)

plt.savefig("correlation.png")
# Adjust layout
plt.tight_layout()
plt.show()
```
