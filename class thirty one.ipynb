```python
nonzero_glocuse = df[df['Glucose']!=0]
```

```python
nonzero_bloodpressure = nonzero_glocuse[nonzero_glocuse['BloodPressure'] != 0]
nonzero_bloodpressure
```

```python
nonzero_skinthickness = nonzero_bloodpressure[nonzero_bloodpressure['SkinThickness']!=0]
nonzero_skinthickness
```

```python
nonzero_bmi = nonzero_skinthickness[nonzero_skinthickness['BMI']!= 0]
nonzero_bmi
```

```python
cleandf = nonzero_bmi[nonzero_bmi['Insulin']!=0]
cleandf
```

```python
cleandf.describe()
```

```python
import numpy as np

# --- Step 1: Replace physiologically implausible values with NaN ---

# BloodPressure: diastolic < 40 mmHg is implausible
cleandf.loc[cleandf['BloodPressure'] < 40, 'BloodPressure'] = np.nan

# Insulin: values > 300 µU/mL are extremely unlikely in fasting state
cleandf.loc[cleandf['Insulin'] > 300, 'Insulin'] = np.nan

# SkinThickness: values < 10 mm may indicate missing data
cleandf.loc[cleandf['SkinThickness'] < 10, 'SkinThickness'] = np.nan

# Glucose: values < 50 mg/dL are too low for typical records
cleandf.loc[cleandf['Glucose'] < 50, 'Glucose'] = np.nan

# BMI: values < 10 or > 70 are unrealistic
cleandf.loc[(cleandf['BMI'] < 10) | (cleandf['BMI'] > 70), 'BMI'] = np.nan

# --- Step 2: Detect and handle outliers using IQR method ---
cols_to_check = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
                 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

for col in cols_to_check:
    Q1 = cleandf[col].quantile(0.25)
    Q3 = cleandf[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Mark extreme outliers as NaN
    cleandf.loc[(cleandf[col] < lower_bound) | (cleandf[col] > upper_bound), col] = np.nan

# --- Step 3: (Optional) Handle NaN values ---
# Option 1: Drop rows with NaN
# cleandf.dropna(inplace=True)

# Option 2: Impute missing values with median
# cleandf.fillna(cleandf.median(), inplace=True)
```
